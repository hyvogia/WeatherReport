```python
import numpy as np
```

```python
pokerhand_array = np.loadtxt("pokerhand.txt",delimiter=",")
```

```python
#First Part
def ace_percentage(arr):
    ace = []
    for i in range(len(pokerhand_array)):
        list = [arr[i,1], arr[i,3], arr[i,5], arr[i,7], arr[i,9]]
        count = ["A" for num in list if num == 1]
        if len(count) >= 2: ace.append("ace")
    return len(ace) / 10
```

```python
output = f"A pair of aces came up {ace_percentage(pokerhand_array)}% of the time out of 1000 hands"
```

```python
print(output)
```

```python
#Second Part
def find_aces(arr):
    count = ["A" for num in arr.flatten() if num == 1]
    return len(count)
```

```python
first_card = pokerhand_array[:, 1] == 1
```

```python
river_card = pokerhand_array[first_card, 9]
```

```python
output = f"Out of 1000 hands, when an ace came up first, the river card was an ace {find_aces(river_card)} out of {find_aces(first_card)} times"
```

```python
print(output)
```
