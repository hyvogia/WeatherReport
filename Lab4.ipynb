```python
data = pd.DataFrame(list(collection.find()))
```

```python
def search_content(s):
    search_result = list(s)
    for s in search_result:
        print(s["content"])
```

```python
def search_title(s):
    search_result = list(s)
    for s in search_result:
        print("The article: " + s["title"])
```

```python
def menu():
    temp_col = collection.find({})
    print("Topics:")
    search_content(temp_col)
    print("Press [E] Exit")
```

```python
def confirm(c):
    for i in range(len(data.content)):
        if data.content[i] == c: 
            return c
```

```python
def program():
    menu()
    option = input("Select a category: ")
    while option != " ":
        if option == confirm(option):
            search_title(collection.find({"content":option},{"_id":0, "title":1}))
            option = input("Select a category: ")
        elif option.upper() == "E":
            print("Bye. Have a great day")
            break
        else:
            print("Invalid Category")
            option = input("Select a category: ")
```

```python
program()
```
